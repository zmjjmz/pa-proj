```python
# Initalize randomly for now
# constants: w, phi, n oscillators

import numpy as np
from numpy import random as rnp

class CPG:
    
    def __init__(self, n, w, osc_class, phi, a, theta, ampl, ampl_dot):
        self.n = n
        # will not be updated
        self.w = w
        self.osc_class = osc_class # 0 for body oscillator, 1 for limb oscillator
        self.phi = phi
        self.a = a
        # Initial vectors, will be updated as simulation goes
        self.theta = theta
        self.ampl = ampl
        self.ampl_dot = ampl_dot
        
        # From table S1 in the supplemental materials
        # each c parameters is [body,limb]
        self.cv0 = [0.3, 0.0]
        self.cv1 = [0.2, 0.2]
        self.cR0 = [0.196,0.131]
        self.cR1 = [0.065,0.131]
        #[[dbodylow,dbodyhigh],[dlimblow,dlimbhigh]]
        self.d_params = [[1,5],[1,3]]
        
    def drive_func(self, d, i):
        """ Given the drive and the oscillator index (to determine if it's a limb or body), get the drive """
        is_limb_osc = self.osc_class[i]
        if d >= self.d_params[is_limb_osc][0] or d <= self.d_params[is_limb_osc][1]:  
            R = self.cR1[is_limb_osc]*d + self.cR0[is_limb_osc]
            v = self.cv1[is_limb_osc]*d + self.cv0[is_limb_osc]
        else:
            R = 0
            v = 0
        
        return R, v

    def step_theta(self, v):
        """ Updates the phase of the oscillators given v"""
        theta_dot = np.zeros(self.n) #  can't vectorize due to convolution
        for i in range(self.n):
            theta_dot[i] = 2*np.pi*v[i]
            theta_dot[i] += sum([self.ampl[j]*np.sin(self.theta[j] - self.theta[i] - self.phi[i][j]) for j in range(self.n)])

        new_theta = self.theta + theta_dot
        return new_theta

    def step_ampl_dot(self, R):
        """ Steps the 2nd order change in amplitude, with the first order as a parameter """
        ampl_dot_dot = self.a*(self.a*(R-self.ampl) - self.ampl_dot)
        new_ampl_dot = self.ampl_dot + ampl_dot_dot
        return new_ampl_dot

    def oscillator_signal(self):
        return self.ampl*(1+np.cos(self.theta))
    
    def step(self, d):
        """ Step the parameters given a drive d """
        R, v = zip(*[self.drive_func(d,i) for i in range(self.n)]) 
        # for the uninitiated
        # zip takes n lists and 'fuses' them elementwise, e.g. zip([1,2,3], [3,4,5]) -> [(1,3),(2,4),(3,5)]
        # the * unpacks a list into arguments to a function, and zip 'fuses' the nth element of each into a list which means you'd get essentially the reverse of normal zip
        # in the brackets is a list comprehension -- it's essentially an expression that creates an anonymous list by evaluating the given expression
        # this gives us R and v as vectors of length n
        # I would recommend restructuring how drive_func works in the C# version of this to not have to do this
        
        R = np.array(R)
        v = np.array(v)
        
        new_theta = self.step_theta(v)
        new_ampl_dot = self.step_ampl_dot(R)
        new_ampl = self.ampl + new_ampl_dot
        
        self.signal = self.oscillator_signal()
        
        # Do stuff with the values here
        
        self.theta = new_theta
        self.ampl_dot = new_ampl_dot
        self.ampl = new_ampl
        
    def run(self, d, t):
        """ Runs t steps with drive d """
        for s in range(t):
            self.step(d)    
    
```

```python
# Test it here

n = 3

w = rnp.rand(n,n)
phi = rnp.rand(n,n)
a = rnp.rand(n)
osc_class = rnp.randint(2,size=n)

# initial states, also random for now
theta = rnp.rand(n)
ampl = rnp.rand(n) # r
ampl_dot = rnp.rand(n) # r_dot

cpg_test = CPG(n, w, osc_class, phi, a, theta, ampl, ampl_dot)

cpg_test.run(1, 10)
print(cpg_test.signal)
```
